# Import Libraries for calculations 

```python
import numpy as np
from IPython.display import clear_output
```

# 1. Define sigmoid

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

# 2. Let's define some helper functions to calculate netH and netY
which is nothing but linear combination on inputs x1,x2 and bais b1 and b2.
we can write equation as fllows (here net means total how perticular neuron is affecting other neuron) 

![](https://miro.medium.com/max/470/1*w2pRQKB05bynMRIyOFPuuA.png)

```python
def calculate_neth(wieghts,wi,b,bi=0):
    neth=(wieghts[wi]*x1)+(wieghts[wi+1]*x2)+b[bi]
    return neth
# (wieghts[4]*Outh1)+(wieghts[5]*Outh2)+b[1]
def calculate_nety(wieghts,wi,out,out2,b,bi=1):
    nety=(wieghts[wi]*out)+(wieghts[wi+1]*out2)+b[bi]
    return nety
```

# 3. After Calculating net output of perticular neuron
**After Calculating netH output ofperticular neuron we simply apply sigmoid function or any other activation function to that output 
 and we calculae netY and then then we calculate error by below equation**

# The Forward Pass

<img src='./nnnn.jpg'>

**in Forward Pass we simply calculate net and out for hidden and output nuerons ( where out is output of activation function )**

**at end of the Pass we calculate Mean squared error
by following equation**
![](https://s0.wp.com/latex.php?latex=E_%7Btotal%7D+%3D+%5Csum+%5Cfrac%7B1%7D%7B2%7D%28target+-+output%29%5E%7B2%7D&bg=ffffff&fg=404040&s=0&c=20201002)
**where target is 0.01 and 0.95 in our case**

```python
histry = []
Net = []
Out = []
def fwdpass(wieghts,x1,x2,b,TargetY1,TargetY2):
    clear_output(wait=True)
    #Neth1 = (wieghts[0]*x1)+(wieghts[1]*x2)+b[0]
    Neth1 = calculate_neth(wieghts,0,b)
    Outh1 = sigmoid(Neth1)
    Net.append(Neth1)
    Out.append(Outh1)
    
    #Neth2 = (wieghts[2]*x1)+(wieghts[3]*x2)+b[0]
    Neth2 = calculate_neth(wieghts,2,b)
    Outh2 = sigmoid(Neth2)
    Net.append(Neth2)
    Out.append(Outh2)
    
    #calculate_nety(wieghts,wi,out,out2,b,bi=1):
    Nety1 = calculate_nety(wieghts,4,Outh1,Outh2,b)
    Outy1 = sigmoid(Nety1)
    Net.append(Nety1)
    Out.append(Outy1)

    print("Output of y1 is:",Outy1)
    Nety2 = calculate_nety(wieghts,6,Outh1,Outh2,b)
    #(wieghts[6]*Outh1)+(wieghts[7]*Outh2)+b[1]
    Outy2 = sigmoid(Nety2)
    Net.append(Nety2)
    Out.append(Outy2)

    print("Output of y2 is:",Outy2)
    # Error Calculation (squared)
    Ey1 = ((TargetY1-Outy1)**2)/2
    Ey2 = ((TargetY2-Outy2)**2)/2
    Etotal = Ey1 + Ey2
    histry.append(Etotal)
    print("Total Error is:",Etotal)
    return wieghts,Neth1,Outh1,Neth2,Outh2,Nety1,Outy1,Nety2,Outy2,Etotal
```

**The learning rate is a hyperparameter that controls how much to change the wieghts in response to the estimated error each time the model weights are updated If your learning rate is set too low, training will progress very slowly as you are making very tiny updates to the weights in your network. However, if your learning rate is set too high, it can cause undesirable divergent behavior in your loss function.**

```python
adjustedW = []
def newwieghts(W,dw):
    eta= 0.5
    nw= W-eta*dw
    adjustedW.append(nw)
    return nw
```

![](https://s0.wp.com/latex.php?latex=%5Cfrac%7B%5Cpartial+E_%7Btotal%7D%7D%7B%5Cpartial+w_%7B5%7D%7D+%3D+%5Cfrac%7B%5Cpartial+E_%7Btotal%7D%7D%7B%5Cpartial+out_%7Bo1%7D%7D+%2A+%5Cfrac%7B%5Cpartial+out_%7Bo1%7D%7D%7B%5Cpartial+net_%7Bo1%7D%7D+%2A+%5Cfrac%7B%5Cpartial+net_%7Bo1%7D%7D%7B%5Cpartial+w_%7B5%7D%7D&bg=ffffff&fg=404040&s=0&c=20201002)

```python
def get_dw(i,j,k):
    return i * j* k
```

```python
def partialDerivative(Targeti, Outyi):
    return -Targeti + Outyi
def partialDerivative2(Outyi):
    return Outyi * (1 - Outyi)
#dEy2Outh2= (-TargetY2 + Outy2) * (Outy2 * (1 - Outy2)) * W[7]
def partialDerivative3(J,K,W,i):
    return J * K * W[i]
```

```python
def pder(x1,x2,TargetY1,TargetY2,W,Neth1,Outh1,Neth2,Outh2,Nety1,Outy1,Nety2,Outy2,Etotal):
    # ∂Etotal/∂w56 = ∂Etotal/∂Outy1 * ∂Outy1/∂Nety1 * ∂Nety1/∂w56
    dEtOy1 = partialDerivative(TargetY1, Outy1)
    dOy1Ny1 = partialDerivative2(Outy1)
    dNy1w5 = Outh1
    dNy1w6 = Outh2
    # ∂Etotal/∂w78 = ∂Etotal/∂Outy1 * ∂Outy1/∂Nety1 * ∂Nety1/∂w78
    dEtOy2 = partialDerivative(TargetY2, Outy2)
    dOy2Ny2 = partialDerivative2(Outy2)
    dNy1w7 = Outh1
    dNy1w8 = Outh2
    # ∂Etotal/∂w12 = ∂Etotal/∂Outh1 * ∂Outh1/∂Neth1 * ∂Neth1/∂w12
    # ∂Etotal/∂Outh1= ∂Ey1/∂Outh1 + ∂Ey2/∂Outh1
    # ∂Ey1/∂Outh1 = ∂Ey1/∂Outy1 * ∂Outy1/∂Nety1 * ∂Nety1/∂Outh1
    dEy1Outh1= partialDerivative3(partialDerivative(TargetY1 ,Outy1),partialDerivative2(Outy1), W,4)
    # ∂Ey2/∂Outh1 = ∂Ey2/∂Outy2 * ∂Outy2/∂Nety2 * ∂Nety2/∂Outh1
    dEy2Outh1= partialDerivative3(partialDerivative(TargetY2 , Outy2),partialDerivative2(Outy2), W,6)
    dOh1Nh1 = partialDerivative2(Outh1)
    dNh1w1 = x1
    dNh1w2 = x2
    # ∂Etotal/∂w34 = ∂Etotal/∂Outh2 * ∂Outh2/∂Neth2 * ∂Neth2/∂w34
    # ∂Etotal/∂Outh2= ∂Ey1/∂Outh2 + ∂Ey2/∂Outh2
    # ∂Ey1/∂Outh2 = ∂Ey1/∂Outy1 * ∂Outy1/∂Nety1 * ∂Nety1/∂Outh2
    dEy1Outh2= partialDerivative3(partialDerivative(TargetY1, Outy1),partialDerivative2(Outy1), W,5)
    # ∂Ey2/∂Outh2 = ∂Ey2/∂Outy2 * ∂Outy2/∂Nety2 * ∂Nety2/∂Outh2
    dEy2Outh2= partialDerivative3(partialDerivative(TargetY2 ,Outy2),partialDerivative2(Outy2), W,7)
    dOh2Nh2 = partialDerivative2(Outh2)
    dNh2w3 = x1
    dNh2w4 = x2
    return dEtOy1,dOy1Ny1,dNy1w5,dNy1w6,dEtOy2,dOy2Ny2,dNy1w7,dNy1w8,dEy1Outh1,dEy2Outh1,dOh1Nh1,dNh1w1,dNh1w2,dEy1Outh2,dEy2Outh2,dOh2Nh2,dNh2w3,dNh2w4
```

# The Backwards Pass
Our goal with backpropagation is to update each of the weights in the network so that they cause the actual output to be closer the target output, thereby minimizing the error for each output neuron and the network as a whole.

![](https://matthewmazur.files.wordpress.com/2015/03/nn-calculation.png)

```python
def backp(W,dEtOy1,dOy1Ny1,dNy1w5,dNy1w6,dEtOy2,dOy2Ny2,dNy1w7,dNy1w8,dEy1Outh1,dEy2Outh1,dOh1Nh1,dNh1w1,dNh1w2,dEy1Outh2,dEy2Outh2,dOh2Nh2,dNh2w3,dNh2w4,x1,x2,b,TargetY1,TargetY2):
    dw5 = get_dw(dEtOy1 ,dOy1Ny1 , dNy1w5)
    dw6 = get_dw(dEtOy1 , dOy1Ny1 , dNy1w6)
    dw7 = get_dw(dEtOy2 , dOy2Ny2 , dNy1w7)
    dw8 = get_dw(dEtOy2 , dOy2Ny2 , dNy1w8)
    
    dw1 = (dEy1Outh1 + dEy2Outh1) * dOh1Nh1 * dNh1w1
    dw2 = (dEy1Outh1 + dEy2Outh1) * dOh1Nh1 * dNh1w2
    dw3 = (dEy1Outh2 + dEy2Outh2) * dOh2Nh2 * dNh2w3
    dw4 = (dEy1Outh2 + dEy2Outh2) * dOh2Nh2 * dNh2w4

    w1= [newwieghts(W[0],dw1),newwieghts(W[1],dw2),newwieghts(W[2],dw3),newwieghts(W[3],dw4),newwieghts(W[4],dw5),newwieghts(W[5],dw6),newwieghts(W[6],dw7),newwieghts(W[7],dw8)]

    print(w1)
    
    return fwdpass(w1,x1,x2,b,TargetY1,TargetY2)
```

# Init weights,Baises and Targets

```python
x1 = 0.05 #float(input("Enter value of X1: "))
x2 = 0.10 #float(input("Enter value of X2: "))
TargetY1= 0.01 
TargetY2= 0.95
W = [0.15,0.20,0.25,0.30,0.40,0.45,0.50,0.55]
# W = []
# for i in range(0,8):
#     w= np.random.uniform(0,1)
#     W.append(w)
b = [0.35,0.60] 
```

# for 1 Forward pass and calculation of derivatives

```python
wieghts,Neth1,Outh1,Neth2,Outh2,Nety1,Outy1,Nety2,Outy2,Etotal= fwdpass(W,x1,x2,b,TargetY1,TargetY2)
dEtOy1,dOy1Ny1,dNy1w5,dNy1w6,dEtOy2,dOy2Ny2,dNy1w7,dNy1w8,dEy1Outh1,dEy2Outh1,dOh1Nh1,dNh1w1,dNh1w2,dEy1Outh2,dEy2Outh2,dOh2Nh2,dNh2w3,dNh2w4 = pder(x1,x2,TargetY1,TargetY2,W,Neth1,Outh1,Neth2,Outh2,Nety1,Outy1,Nety2,Outy2,Etotal)
```

# using The rectified linear unit
**relu activation function or ReLU for short is a piecewise linear function that will output the input directly if it is positive, otherwise, it will output zero.**

# g ( z ) = max { 0 , z }


![](https://miro.medium.com/max/714/1*oePAhrm74RNnNEolprmTaQ.png)

```python
def relu(L):
    if L<=0:
        return 0
    else:
        return L
```

```python
#overwrite function
def partialDerivative2(Outyi):
    if Outyi<=0:
        return 0
    else:
        return 1
```

```python
histryRelu = []
NetRelu = []
OutRelu = []
def FwdpassRelu(wieghts,x1,x2,b,TargetY1,TargetY2):
    clear_output(wait=True)
    #Neth1 = (wieghts[0]*x1)+(wieghts[1]*x2)+b[0]
    Neth1 = calculate_neth(wieghts,0,b)
    Outh1 = relu(Neth1)
    NetRelu.append(Neth1)
    OutRelu.append(Outh1)
    
    #Neth2 = (wieghts[2]*x1)+(wieghts[3]*x2)+b[0]
    Neth2 = calculate_neth(wieghts,2,b)
    Outh2 = relu(Neth2)
    NetRelu.append(Neth2)
    OutRelu.append(Outh2)
    
    #calculate_nety(wieghts,wi,out,out2,b,bi=1):
    Nety1 = calculate_nety(wieghts,4,Outh1,Outh2,b)
    Outy1 = relu(Nety1)
    NetRelu.append(Nety1)
    OutRelu.append(Outy1)

    print("Output of y1 is:",Outy1)
    Nety2 = calculate_nety(wieghts,6,Outh1,Outh2,b)
    #(wieghts[6]*Outh1)+(wieghts[7]*Outh2)+b[1]
    Outy2 = relu(Nety2)
    NetRelu.append(Nety2)
    OutRelu.append(Outy2)

    print("Output of y2 is:",Outy2)
    # Error Calculation (squared)
    Ey1 = ((TargetY1-Outy1)**2)/2
    Ey2 = ((TargetY2-Outy2)**2)/2
    Etotal = Ey1 + Ey2
    histryRelu.append(Etotal)
    print("Total Error is:",Etotal)
    return wieghts,Neth1,Outh1,Neth2,Outh2,Nety1,Outy1,Nety2,Outy2,Etotal
```

# Binary Step Function

![](https://miro.medium.com/max/381/1*K9QJmeG33SvQeJgPa52mmQ.png)
**Binary step function is a threshold-based activation function which means after a certain threshold neuron is activated and below the said threshold neuron is deactivated. In the above graph, the threshold is zero. This activation function can be used in binary classifications as the name suggests, however it can not be used in a situation where you have multiple classes to deal with.**

```python
#hevysidestep function = bsf
def HSSP(X):
    if X>0:
        return 1.0
    else:
        return 0.0
```

```python
#overwrite function
def partialDerivative2(Outyi):
    if Outyi>=0:
        return 1.0
    else:
        return 0.0
```

```python
histrybsf = []
Netbsf = []
Outbsf = []
def Fwdpassbsf(wieghts,x1,x2,b,TargetY1,TargetY2):
    clear_output(wait=True)
    #Neth1 = (wieghts[0]*x1)+(wieghts[1]*x2)+b[0]
    Neth1 = calculate_neth(wieghts,0,b)
    Outh1 = HSSP(Neth1)
    Netbsf.append(Neth1)
    Outbsf.append(Outh1)
    
    #Neth2 = (wieghts[2]*x1)+(wieghts[3]*x2)+b[0]
    Neth2 = calculate_neth(wieghts,2,b)
    Outh2 = HSSP(Neth2)
    Netbsf.append(Neth2)
    Outbsf.append(Outh2)
    
    #calculate_nety(wieghts,wi,out,out2,b,bi=1):
    Nety1 = calculate_nety(wieghts,4,Outh1,Outh2,b)
    Outy1 = HSSP(Nety1)
    Netbsf.append(Nety1)
    Outbsf.append(Outy1)

    print("Output of y1 is:",Outy1)
    Nety2 = calculate_nety(wieghts,6,Outh1,Outh2,b)
    #(wieghts[6]*Outh1)+(wieghts[7]*Outh2)+b[1]
    Outy2 = HSSP(Nety2)
    Netbsf.append(Nety2)
    Outbsf.append(Outy2)

    print("Output of y2 is:",Outy2)
    # Error Calculation (squared)
    Ey1 = ((TargetY1-Outy1)**2)/2
    Ey2 = ((TargetY2-Outy2)**2)/2
    Etotal = Ey1 + Ey2
    histrybsf.append(Etotal)
    print("Total Error is:",Etotal)
    return wieghts,Neth1,Outh1,Neth2,Outh2,Nety1,Outy1,Nety2,Outy2,Etotal
```

```python
def backp(W,dEtOy1,dOy1Ny1,dNy1w5,dNy1w6,dEtOy2,dOy2Ny2,dNy1w7,dNy1w8,dEy1Outh1,dEy2Outh1,dOh1Nh1,dNh1w1,dNh1w2,dEy1Outh2,dEy2Outh2,dOh2Nh2,dNh2w3,dNh2w4,x1,x2,b,TargetY1,TargetY2):
    dw5 = get_dw(dEtOy1 ,dOy1Ny1 , dNy1w5)
    dw6 = get_dw(dEtOy1 , dOy1Ny1 , dNy1w6)
    dw7 = get_dw(dEtOy2 , dOy2Ny2 , dNy1w7)
    dw8 = get_dw(dEtOy2 , dOy2Ny2 , dNy1w8)
    
    dw1 = (dEy1Outh1 + dEy2Outh1) * dOh1Nh1 * dNh1w1
    dw2 = (dEy1Outh1 + dEy2Outh1) * dOh1Nh1 * dNh1w2
    dw3 = (dEy1Outh2 + dEy2Outh2) * dOh2Nh2 * dNh2w3
    dw4 = (dEy1Outh2 + dEy2Outh2) * dOh2Nh2 * dNh2w4

    w1= [newwieghts(W[0],dw1),newwieghts(W[1],dw2),newwieghts(W[2],dw3),newwieghts(W[3],dw4),newwieghts(W[4],dw5),newwieghts(W[5],dw6),newwieghts(W[6],dw7),newwieghts(W[7],dw8)]

    print(w1)
    wieghts,Neth1,Outh1,Neth2,Outh2,Nety1,Outy1,Nety2,Outy2,Etotal=Fwdpassbsf(w1,x1,x2,b,TargetY1,TargetY2)
    return wieghts,Neth1,Outh1,Neth2,Outh2,Nety1,Outy1,Nety2,Outy2,Etotal
```

# tanh
![](https://miro.medium.com/max/381/1*YRc3zhQginJokn2H0737dA.jpeg)
**tanh function (also known as hyperbolic tangent) is almost like the sigmoid function but slightly better than that since it’s output ranges between -1 and 1 allowing negative outputs.**

```python
def tanH(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
#overwrite function
def partialDerivative2(x):
    return 1 - ((np.exp(x) - np.exp(-x))**2) / ((np.exp(x) + np.exp(-x))**2)
```

```python
histrytanH = []
NettanH = []
OuttanH = []
def FwdpasstanH(wieghts,x1,x2,b,TargetY1,TargetY2):
    clear_output(wait=True)
    #Neth1 = (wieghts[0]*x1)+(wieghts[1]*x2)+b[0]
    Neth1 = calculate_neth(wieghts,0,b)
    Outh1 = tanH(Neth1)
    NettanH.append(Neth1)
    OuttanH.append(Outh1)
    #Neth2 = (wieghts[2]*x1)+(wieghts[3]*x2)+b[0]
    Neth2 = calculate_neth(wieghts,2,b)
    Outh2 = tanH(Neth2)
    NettanH.append(Neth2)
    OuttanH.append(Outh2)
    
    #calculate_nety(wieghts,wi,out,out2,b,bi=1):
    Nety1 = calculate_nety(wieghts,4,Outh1,Outh2,b)
    Outy1 = tanH(Nety1)
    NettanH.append(Nety1)
    OuttanH.append(Outy1)
    
    print("Output of y1 is:",Outy1)
    Nety2 = calculate_nety(wieghts,6,Outh1,Outh2,b)
    #(wieghts[6]*Outh1)+(wieghts[7]*Outh2)+b[1]
    Outy2 = tanH(Nety2)
    NettanH.append(Nety2)
    OuttanH.append(Outy2)
    print("Output of y2 is:",Outy2)
    # Error Calculation (squared)
    Ey1 = ((TargetY1-Outy1)**2)/2
    Ey2 = ((TargetY2-Outy2)**2)/2
    Etotal = Ey1 + Ey2
    histrytanH.append(Etotal)
    print("Total Error is:",Etotal)
    return wieghts,Neth1,Outh1,Neth2,Outh2,Nety1,Outy1,Nety2,Outy2,Etotal
```

```python
def backp(W,dEtOy1,dOy1Ny1,dNy1w5,dNy1w6,dEtOy2,dOy2Ny2,dNy1w7,dNy1w8,dEy1Outh1,dEy2Outh1,dOh1Nh1,dNh1w1,dNh1w2,dEy1Outh2,dEy2Outh2,dOh2Nh2,dNh2w3,dNh2w4,x1,x2,b,TargetY1,TargetY2):
    dw5 = get_dw(dEtOy1 ,dOy1Ny1 , dNy1w5)
    dw6 = get_dw(dEtOy1 , dOy1Ny1 , dNy1w6)
    dw7 = get_dw(dEtOy2 , dOy2Ny2 , dNy1w7)
    dw8 = get_dw(dEtOy2 , dOy2Ny2 , dNy1w8)
    
    dw1 = (dEy1Outh1 + dEy2Outh1) * dOh1Nh1 * dNh1w1
    dw2 = (dEy1Outh1 + dEy2Outh1) * dOh1Nh1 * dNh1w2
    dw3 = (dEy1Outh2 + dEy2Outh2) * dOh2Nh2 * dNh2w3
    dw4 = (dEy1Outh2 + dEy2Outh2) * dOh2Nh2 * dNh2w4
    w1= [newwieghts(W[0],dw1),newwieghts(W[1],dw2),newwieghts(W[2],dw3),newwieghts(W[3],dw4),newwieghts(W[4],dw5),newwieghts(W[5],dw6),newwieghts(W[6],dw7),newwieghts(W[7],dw8)]
    print(w1)
    wieghts,Neth1,Outh1,Neth2,Outh2,Nety1,Outy1,Nety2,Outy2,Etotal=FwdpasstanH(w1,x1,x2,b,TargetY1,TargetY2)
    return wieghts,Neth1,Outh1,Neth2,Outh2,Nety1,Outy1,Nety2,Outy2,Etotal
```
